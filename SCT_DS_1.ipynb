**Task 1-** Sales Distribution Analysis

Internship Domain: Data Science

**Dataset:** Sales by Product Dataset

**Objective:** To analyze and visualize product-wise sales distribution using a bar chart and histogram

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create the sales dataset
data = {
    "Product": ["Laptop", "Phone", "Tablet", "Monitor", "Keyboard", "Mouse", "Headset", "Pendrive", "Charger"],
    "Sales": [120, 300, 180, 90, 60, 75, 50, 85, 150]
}

df = pd.DataFrame(data)


# Step 2: Create the bar chart
plt.figure(figsize=(8, 5))
plt.bar(df["Product"], df["Sales"])
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Sales Distribution by Product")

# Step 3: Save the chart
plt.savefig("sales_by_product.png", bbox_inches="tight")

# Show the chart
plt.show()
```

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Create the sales dataset
data = {
    "Product": ["Laptop", "Phone", "Tablet", "Monitor", "Keyboard", "Mouse", "Headset", "Pendrive", "Charger"],
    "Sales": [120, 300, 180, 90, 60, 75, 50, 85, 150]
}

df = pd.DataFrame(data)


# Create histogram for Sales
plt.figure(figsize=(8, 5))
plt.hist(df["Sales"], bins=5)
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.title("Sales Distribution Histogram")

# Save the histogram
plt.savefig("sales_histogram.png", bbox_inches="tight")

# Show the histogram
plt.show()
```
